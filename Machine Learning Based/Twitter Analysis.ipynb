# Data Visualization 

```python
# Calculating company-wise aggregates
company_stats = twi.groupby('Company').agg({'Retweets': 'sum', 'Likes': 'sum', 'Hashtags': 'count'})

# Plotting individual companies
company_stats.plot(kind='bar', subplots=True, layout=(1, 3), figsize=(15, 5))
plt.show()
```

```python

# Group the data by the 'Company' column and sum the 'Likes' for each company
likes_by_company = twi.groupby('Company')['Likes'].sum()

# Find the company with the most likes
company_with_most_likes = likes_by_company.idxmax()
most_likes = likes_by_company.max()

# Create a bar plot to compare likes among all companies
plt.figure(figsize=(10, 6))
likes_by_company.plot(kind='bar')
plt.title('Likes Comparison by Company')
plt.xlabel('Company')
plt.ylabel('Likes')
plt.xticks(rotation=40)
plt.tight_layout()

plt.show()

print(f"The company with the most likes is {company_with_most_likes} with {most_likes} likes.")
```

```python
# Group the data by 'Company' and sum the 'Retweets'
retweets_by_company = twi.groupby('Company')['Retweets'].sum()

# Sort companies by retweets in descending order
sorted_retweets = retweets_by_company.sort_values(ascending=False)

# Plot a bar chart
plt.figure(figsize=(10, 6))
sorted_retweets.plot(kind='bar', color='skyblue')
plt.title('Retweets by Company')
plt.xlabel('Company')
plt.ylabel('Retweets')
plt.xticks(rotation=45)
plt.show()
```

# Regression Based Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Read the Twitter data from the CSV file
twi = pd.read_csv(r"C:\Users\Asus\Downloads\twitter.csv")

# Create a label encoder to convert company names to numerical labels
label_encoder = LabelEncoder()
twi['company_label'] = label_encoder.fit_transform(twi['Company'])

# Define the feature columns (likes and retweets)
X = twi[['Likes', 'Retweets']]

# Define the target variable (company labels)
y = twi['company_label']

# Create and fit a logistic regression model
classification_model = LogisticRegression()
classification_model.fit(X, y)

# Predict the company labels based on likes and retweets
predicted_labels = classification_model.predict(X)

# Calculate the accuracy of the model
accuracy = accuracy_score(y, predicted_labels)

# Inverse transform the predicted labels to get company names
predicted_companies = label_encoder.inverse_transform(predicted_labels)

# Add the predicted companies to the DataFrame
twi['predicted_company'] = predicted_companies

# Group the data by 'Company' to calculate the total likes and retweets for each company
company_likes_retweets = twi.groupby('Company')[['Likes', 'Retweets']].sum()

# Calculate the combined likes and retweets for each company
company_likes_retweets['combined_likes_retweets'] = company_likes_retweets['Likes'] + company_likes_retweets['Retweets']

# Find the company with the highest combined likes and retweets
highest_likes_retweets_company = company_likes_retweets.idxmax()['combined_likes_retweets']

# Create a line plot to visualize the relationship between the company and combined likes and retweets
plt.figure(figsize=(10, 6))
sns.lineplot(data=company_likes_retweets, x=company_likes_retweets.index, y='combined_likes_retweets', label='Combined Likes and Retweets')
plt.title('Companies with Highest Combined Likes and Retweets')
plt.xlabel('Company')
plt.ylabel('Combined Likes and Retweets')
plt.xticks(rotation=45)
plt.axvline(x=highest_likes_retweets_company, color='red', linestyle='--', label=f'Highest: {highest_likes_retweets_company}')
plt.legend()
plt.show()
```

# Decision Tree Based Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Read the Twitter data from the CSV file
twi = pd.read_csv(r"C:\Users\Asus\Downloads\twitter.csv")

# Create a label encoder to convert company names to numerical labels
label_encoder = LabelEncoder()
twi['company_label'] = label_encoder.fit_transform(twi['Company'])

# Define the feature columns (likes and retweets)
X = twi[['Likes', 'Retweets']]

# Define the target variable (company labels)
y = twi['company_label']

# Create and fit a Decision Tree Classifier model
classification_model = DecisionTreeClassifier(random_state=42)
classification_model.fit(X, y)

# Predict the company labels based on likes and retweets
predicted_labels = classification_model.predict(X)

# Inverse transform the predicted labels to get company names
predicted_companies = label_encoder.inverse_transform(predicted_labels)

# Add the predicted companies to the DataFrame
twi['predicted_company'] = predicted_companies

# Group the data by 'Company' to calculate the total likes and retweets for each company
company_likes_retweets = twi.groupby('Company')[['Likes', 'Retweets']].sum()

# Calculate the combined likes and retweets for each company
company_likes_retweets['combined_likes_retweets'] = company_likes_retweets['Likes'] + company_likes_retweets['Retweets']

# Find the company with the highest combined likes and retweets
highest_likes_retweets_company = company_likes_retweets.idxmax()['combined_likes_retweets']

# Sort the companies by combined likes and retweets in descending order for the bar plot
company_likes_retweets = company_likes_retweets.sort_values(by='combined_likes_retweets', ascending=False)

# Create a colorful bar plot to visualize the company with the highest combined likes and retweets
plt.figure(figsize=(10, 6))
colors = sns.color_palette("viridis", len(company_likes_retweets))

ax = sns.barplot(x='combined_likes_retweets', y=company_likes_retweets.index, data=company_likes_retweets, palette=colors)
plt.title('Companies with Highest Combined Likes and Retweets')
plt.xlabel('Combined Likes and Retweets')
plt.ylabel('Company')
plt.axhline(y=highest_likes_retweets_company, color='red', linestyle='--', label=f'Highest: {highest_likes_retweets_company}')
plt.legend()

# Add labels and values to the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_width()):,}', (p.get_width(), p.get_y() + p.get_height() / 2),
                ha='center', va='center', color='black', fontsize=10, fontweight='bold')

plt.show()
```
